```python
# Import statements for the modules used in the code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# setting 
pd.options.mode.chained_assignment = None
```

```python
#read csv file into data frame
# using low_memory flag as getting a warning
df1 = pd.read_csv('loan.csv', low_memory = False)
df1.head()
```

```python
# Check the shape of the data frame to check how many rows and columns are there
df1.shape
```

# Data Cleaning step

### Drop columns with high amount of null values and irrelevant columns

```python
# Check if there are any columns with null values
df1.isnull().any()
```

```python
# Seems like several columns are having null values
# In case there are very high number of null values in a column (more than 40% of the column values) the column wont be useful, 
# hence get the percentage of null values in the columns and drop columns with more than 40% null values
percent_missing = df1.isnull().sum() * 100 / len(df1)
percent_missing
```

```python
cols_to_drop = percent_missing[percent_missing>40].keys()
df2 = df1.drop(columns=cols_to_drop)
df2
```

```python
# we see that now we are left with 54 columns 
# let us now check the percentage of null values in each column again

df2.isnull().sum() * 100 / len(df2)
```

```python
# it is observed that the columns desc has 32% missing values
df2['desc']
# upon checking the content of the desc column it is observed that it does not have data relevant to loans so lets drop it
df2 = df2.drop(columns=['desc'])
```

```python
df2['emp_title']
# upon checking the content of the emp_title column it is observed that it does not have data relevant to loans so lets drop it
df2 = df2.drop(columns=['emp_title'])
```

```python
df2['title']
# upon checking the content of the title column it is observed that it does not have data relevant to loans so lets drop it
df2 = df2.drop(columns=['title'])
```

```python
# from studying the data dictionary it is observed that the url column is unnecessary so drop it
df2 = df2.drop(columns=['url'])
```

```python
# let us remove columns which have non unique values, i.e. all values are same since they will not affect the analysis
nunique = df2.nunique()
columns_to_drop = nunique[nunique == 1].index
df3 = df2.drop(columns_to_drop, axis=1)
df3
```

```python
# the data set contains the data for completed loans, currently running loans and defaulted (charged off) loans
# get rid of rows where the loan_status = current as we cant say if the currently running loan will successfully be completed or not
df4 = df3[df3.loan_status != 'Current']
df4
```

### Imputing Values

```python

df4.shape
df4.isnull().sum() * 100 / len(df4)
```

```python
# Imputing data for emp_length, revol_util, last_pymnt_d, pub_rec_bankruptcies
df4['pub_rec_bankruptcies'].mean()
df4['pub_rec_bankruptcies'].mode()
# the mean and mode for the pub_rec_bankruptcies is similar so let us use mode to impute values
df4.pub_rec_bankruptcies= df4.pub_rec_bankruptcies.fillna(df4['pub_rec_bankruptcies'].mode()[0])
```

```python
# emp lenth is not number type column
df4.emp_length.mode()
# the mean and mode for the emp_length is similar so let us use mode to impute values
df4.emp_length= df4.emp_length.fillna(df4.emp_length.mode()[0])
```

```python
df4['last_pymnt_d']
# last_pymnt_d is date field regarding when the last payment was recieved this field is not important and the column can be dropped 
df4 = df4.drop(columns=['last_pymnt_d'])
```

```python
df4.isnull().sum() * 100 / len(df4)
```

```python
# last_credit_pull_d has very few missing values so drop rows where value is nan
df4 = df4.drop(df4[df4['last_credit_pull_d'].isnull() == True].index, axis=0)
df4.shape
```

```python
df4['revol_util'].unique()
```

```python
# revol util signifies the Revolving line utilization rate, or the amount of credit the borrower is using relative to all available revolving credit.
# since it could be a meaningful column and there are very few values missing drop rows which have null 
df4 = df4.drop(df4[df4['revol_util'].isnull() == True].index, axis=0)
df4.shape
```

```python
df4.isnull().sum() * 100 / len(df4)
```
